# Stage-1 Analysis

Stage 1 analysis focuses on finding correlation of industry employment with total employment.

```python
# Import libraries
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os
from os import *
import sys
sys.path.append("../")
```

```python
# Change directory
def change_dir(newpath):
    get_path = sys.path[0].split("\\")      
    del get_path[-1]                                                
    get_path.append(newpath)                    
    path = "\\".join(get_path)              
    os.chdir(path)
```

## MSA level

```python
# Retrieve industry file filtered for 2019 yearly data
change_dir('output')
df = pd.read_csv("merged_industry_LAUS.csv",index_col=[0])
df = df[df['year']==2019]
df = df[df['period']=='M13']
df = df.fillna(0)
df
```

```python
# Get area code as the index and industries as columns
df1 = df.pivot_table(index='area_code',columns='industry_code',values='industry_employment', aggfunc='mean')
df1 = df1.reset_index()
df1
```

```python
# Find correlations
col_name = list(df1.columns)
col_name.remove('area_code')
lst = []
for i in col_name:
    cor = df1[0].corr(df1[i])
    lst.append(cor)
lst.insert(0,0)
len(lst)
```

```python
# Append the row with correlations to the dataframe
a_series = pd.Series(lst, index = df1.columns)
df2 = df1.append(a_series,ignore_index=True)
df2
```

```python
# Find NaN values column-wise and row-wise
df2 = df2.append(df2.isna().sum(),ignore_index=True)
df2 = df2.append(df1.notnull().sum(),ignore_index=True)
df2
```

```python
# Extract the computation rows at the end of the dataframe
df3 = df2.iloc[-3:]
df3 = df3.transpose()
df3.rename(columns = { 427: 'corrs', 428: 'nans', 429:'MSAs' }, inplace=True)
```

```python
# Add industry names
change_dir('input//industry')
industry_names = pd.read_csv('industry_codes.csv',sep='\t')
merged = df3.merge(industry_names,on='industry_code')
merged 
```

```python
# Check if the industry is a supersector
supersector_names = pd.read_csv('supersector_series.csv',sep='\t')
supersector_names = supersector_names.rename(columns={'supersector_name':'industry_name'})
merged2 = merged.merge(supersector_names,on='industry_name',how='left')
merged2['supersector'] = merged2['supersector_code'].isna()
merged2['supersector'] = merged2['supersector'].replace(False,'Yes')
merged2['supersector'] = merged2['supersector'].replace(True,'No')
merged2
```

```python
# Save output
change_dir('output')
merged2 = merged2.drop(['supersector_code'],axis=1)
merged2.to_csv("stage1.csv",index=False)
```
