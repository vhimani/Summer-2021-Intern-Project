# Employment - Industries

The data is sourced from BLS State and Metro Area Employment, Hours, & Earnings database for time series. The data link is the text data files for all data from 1991 to 2021. The data is updated on a monthly basis.

```python
# Import libraries
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os
from os import *
import sys
sys.path.append("../")
import zipfile
import urllib.request
from shutil import copyfile
import shutil
```

```python
# Change directory
def change_dir(newpath):
    get_path = sys.path[0].split("\\")      
    del get_path[-1]                                                  
    get_path.append(newpath)                    
    path = "\\".join(get_path)              
    os.chdir(path) 
change_dir('input\\industry')
```

```python
# Source all data from the BLS SAE URL for .txt files
url = 'https://download.bls.gov/pub/time.series/sm/sm.data.1.AllData'
urllib.request.urlretrieve(url,'all_data.csv')
all_data = pd.read_csv('all_data.csv', sep='\t',)
all_data.columns = all_data.columns.str.strip()
all_data
```

```python
# Source series data from the BLS SAE URL for .txt files
url = 'https://download.bls.gov/pub/time.series/sm/sm.series'
urllib.request.urlretrieve(url,'series.csv')
series = pd.read_csv('series.csv', sep='\t',)
series.columns = series.columns.str.strip()
series
```

```python
# Merge all_data with series data
all_data_series = pd.merge(all_data, series, on='series_id')
all_data_series
```

```python
# Filter the merged dataset for data type: All Employees, In Thousands and 50 states and DC
all_data_series = all_data_series[all_data_series.data_type_code == 1]
all_data_series = all_data_series[all_data_series.state_code < 57]
all_data_series
```

```python
# Source supersector data from the BLS SAE URL for .txt files
url = 'https://download.bls.gov/pub/time.series/sm/sm.supersector'
urllib.request.urlretrieve(url,'supersector_series.csv')
supersector_series = pd.read_csv('supersector_series.csv', sep='\t',)
supersector_series.columns = supersector_series.columns.str.strip()
supersector_series
```

```python
# Source industry data from the BLS SAE URL for .txt files
url = 'https://download.bls.gov/pub/time.series/sm/sm.industry'
urllib.request.urlretrieve(url,'industry_codes.csv')
industry_codes = pd.read_csv('industry_codes.csv', sep='\t',)
industry_codes.columns = industry_codes.columns.str.strip()
industry_codes
```

```python
# Merge supersector_series with all_data_series 
all_data_supersector = pd.merge(all_data_series, supersector_series, on='supersector_code')
all_data_supersector.columns
```

```python
# Drop irrelevant columns
all_data_supersector = all_data_supersector.drop(['footnote_codes_x','benchmark_year','footnote_codes_y','begin_year', 'begin_period', 'end_year', 'end_period','supersector_name'],axis=1)
all_data_supersector
```

```python
# Filter for supersectors, years after 1990 and unseasonal data
all_data_supersector = all_data_supersector[all_data_supersector.supersector_code != 15]
all_data_supersector = all_data_supersector[all_data_supersector.supersector_code != 30]
all_data_supersector = all_data_supersector[all_data_supersector.supersector_code != 40]
all_data_supersector = all_data_supersector[all_data_supersector.year > 1990]
all_data_supersector = all_data_supersector[all_data_supersector.seasonal != 'S']
all_data_supersector
```

```python
all_data_supersector['industry_code_match'] = all_data_supersector['supersector_code'].astype(str) + '000000'
all_data_supersector['industry_code'] = all_data_supersector['industry_code'].astype(str)
all_data_supersector['industry_code'] = np.where((all_data_supersector.industry_code == '0'),'0000000',all_data_supersector.industry_code)
all_data_supersector = all_data_supersector[all_data_supersector.industry_code == all_data_supersector.industry_code_match]
all_data_supersector.columns
```

```python
# Drop irrelevant columns
all_data_supersector = all_data_supersector.drop(['data_type_code','seasonal','industry_code_match'],axis=1)
all_data_supersector
```

```python
change_dir('output')
all_data_supersector.to_csv('industries.csv')   # Long format file
```

## Industry Dataset for Stage 1 Analysis

```python
# Data preprocessing
all_data_series = all_data_series[all_data_series['year']>2014]
all_data_supersector = pd.merge(all_data_series, supersector_series, on='supersector_code')
all_data_supersector = all_data_supersector[all_data_supersector['seasonal'] != 'S']
all_data_supersector = all_data_supersector.drop(['footnote_codes_x','benchmark_year','footnote_codes_y','begin_year', 'begin_period', 'end_year', 'end_period','seasonal','data_type_code','supersector_name'],axis=1)
all_data_supersector
```

```python
# Filtering industry codes for 3 digit industries
all_data_supersector['industry_code'] = all_data_supersector['industry_code'].astype(str)
all_data_supersector['industry_code'] = np.where((all_data_supersector.industry_code == '0'),'00000000',all_data_supersector.industry_code)
all_data_supersector['industry_last3'] = all_data_supersector['industry_code'].str[5:9]
all_data_supersector = all_data_supersector[(all_data_supersector['industry_last3']=='000') | (all_data_supersector['industry_last3']=='00')]
all_data_supersector = all_data_supersector.drop(['industry_last3'],axis=1)
all_data_supersector.reset_index()
```

```python
# Filter and save for states vs MSA
all_data_supersector_states = all_data_supersector[all_data_supersector['area_code']==0]
all_data_supersector_msa = all_data_supersector[all_data_supersector['area_code']>0]
change_dir('output')
all_data_supersector_states.to_csv('industries_states_s1.csv')
all_data_supersector_msa.to_csv('industries_msa_s1.csv')
```
